### 2.) Plotting Area

this code was used as inspiration:

- https://matplotlib.org/2.0.2/examples/pylab_examples/bar_stacked.html 
- https://matplotlib.org/stable/gallery/statistics/histogram_cumulative.html 

```python
#all labels (after size exclusion)
table_all = pd.read_csv('all_measurements_normalized_image.csv')
table_all_testing = pd.read_csv('testing_all_measurements_normalized_image.csv')
table_all_both = pd.concat([table_all, table_all_testing], ignore_index = True)

#only PL
good_crystals_training = pd.read_csv('dropped_crystal_labels_shape.csv')
good_crystals_testing = pd.read_csv('droppedtesting_crystal_labels_shape.csv')
good_crystals_both = pd.concat([good_crystals_training, good_crystals_testing], ignore_index = True)
```

```python
#area in pixels 
#all labels (after size exclusion)
area1_all = table_all['area']
area2_all = table_all_testing['area']
area_all = table_all_both['area']

#only PL
area1 = good_crystals_training['area']
area2 = good_crystals_testing['area']
area = good_crystals_both['area']
```

```python
#area in um^2
table_all['area [in um^2]'] = (table_all['area']*0.004225)
area1_all_um2 = table_all['area [in um^2]']
table_all.to_csv('all_measurements_normalized_image_um2.csv')

table_all_testing['area [in um^2]'] = (table_all_testing['area']*0.004225)
area2_all_um2 = table_all_testing['area [in um^2]'] 
table_all_testing.to_csv('testing_all_measurements_normalized_image_um2.csv')

table_all_both['area [in um^2]'] = (table_all_both['area']*0.004225)
area_all_um2 = table_all_both['area [in um^2]']
table_all_both.to_csv('all_measurements_both_normalized_image_um2.csv')

good_crystals_training['area [in um^2]'] = (good_crystals_training['area']*0.004225)
area1_um2 = good_crystals_training['area [in um^2]']

good_crystals_testing['area [in um^2]'] = (good_crystals_testing['area']*0.004225)
area2_um2 = good_crystals_testing['area [in um^2]']

good_crystals_both['area [in um^2]'] = (good_crystals_both['area']*0.004225)
area_um2 = good_crystals_both['area [in um^2]'] #$bagshlashmu$
```

```python
median_area_all = np.median(table_all_both['area [in um^2]'])
median_area_good_crystals = np.median(good_crystals_both['area [in um^2]'])
```

```python
font = {'size' : 22}
matplotlib.rc('font',**font)
```

```python
#histogram
# comparison area of all bigger crystal labels and selected good crystal labels
fig, axes = plt.subplots(2, 2, figsize=(20,15))
fig.suptitle('Area', fontsize = 27)

#area measurements all crystal labels using classifier trained on shape
hist1_all_output = axes[0,0].hist(area1_all_um2, bins = 50, label='Area training data', alpha=0.7, color = '#79bafa', edgecolor='black', range = (0,25))
axes[0,0].hist(area2_all_um2, label='Area testing data', bins = 50, alpha=0.7, bottom = hist1_all_output[0], color = '#ff9a00', edgecolor='black', range = (0,25))
axes[0,0].axvline(median_area_all, color = '#ae0001', label = 'Median')
axes[0,0].set_xlabel('area [$\mu m^2$]')
axes[0,0].set_ylabel('count')
axes[0,0].legend()
axes[0,0].set_title('Histogram of AL')
axes[0,0].set_ylim(0,130)

#area measurements good crystal labels using classifier trained on shape
hist1_output = axes[0,1].hist(area1_um2, bins = 50, label='Area training data', alpha=0.7, color = '#79bafa', edgecolor='brown', range = (0,25))
axes[0,1].hist(area2_um2, bins = 50, label='Area testing data', alpha=0.7,  bottom = hist1_output[0], color = '#ff9a00', edgecolor='black', range = (0,25))
axes[0,1].axvline(median_area_good_crystals, color = '#ae0001', label = 'Median')
axes[0,1].set_xlabel('area [$\mu m^2$]')
axes[0,1].set_ylabel('count')
axes[0,1].set_title('Histogram of PL')
axes[0,1].set_ylim(0,130)
axes[0,1].legend()

# cumulative distribution function
axes[1,0].hist(area_all_um2, density=True, histtype='step', cumulative=True, label='AL', color = 'magenta')
axes[1,0].hist(area_um2, density=True, histtype='step', cumulative=True, label='PL', color = 'green')
axes[1,0].grid(True)
axes[1,0].legend(loc='right')
axes[1,0].set_title('CDF of AL and PL')
axes[1,0].set_xlabel('area [$\mu m^2$]')
axes[1,0].set_ylabel('Proportion of occurence')
axes[1,0].set_xlim(0,25)

#boxplot
AL = pd.DataFrame(area_all_um2)
AL['labels'] = ['AL']*len(area_all_um2)
PL = pd.DataFrame(area_um2)
PL['labels'] = ['PL']*len(area_um2)
boxplot_table = pd.concat([AL, PL], ignore_index = True)
boxplot_all = sns.boxplot(x = 'labels', y = 'area [in um^2]',  data = boxplot_table, hue = 'labels', color = '#79bafa', ax = axes[1,1])
axes[1,1].set_ylim([0,25])
axes[1,1].set_ylabel('area [$\mu m^2$]')
axes[1,1].set_title('Boxplot of AL and PL')

plt.tight_layout()
```

```python
#histogram
# comparison area with different classifiers
fig, axes = plt.subplots(2, 3, figsize=(30,20))
fig.suptitle('Area comparison between different classifiers', fontsize = 27)

#area measurements all crystal labels using classifier trained on shape 
hist1_all_output = axes[0,0].hist(area1_all_um2, bins = 50, label='Area training data', alpha=0.7, color = '#79bafa', edgecolor='black', range = (0,25))
axes[0,0].hist(area2_all_um2, label='Area testing data', bins = 50, alpha=0.7, bottom = hist1_all_output[0], color = '#ff9a00', edgecolor='black', range = (0,25))
axes[0,0].axvline(median_area_all, color = '#ae0001', label = 'Median')
axes[0,0].set_xlabel('area [$\mu m^2$]')
axes[0,0].set_ylabel('count')
axes[0,0].legend()
axes[0,0].set_title('Histogram of AL')
axes[0,0].set_ylim(0,130)

#area measurements good crystal labels using classifier trained on shape 
hist1_output = axes[0,1].hist(area1_um2, bins = 50, label='Area training data', alpha=0.7, color = '#79bafa', edgecolor='brown', range = (0,25))
axes[0,1].hist(area2_um2, bins = 50, label='Area testing data', alpha=0.7,  bottom = hist1_output[0], color = '#ff9a00', edgecolor='black', range = (0,25))
axes[0,1].axvline(median_area_good_crystals, color = '#ae0001', label = 'Median')
axes[0,1].set_xlabel('area [$\mu m^2$]')
axes[0,1].set_ylabel('count')
axes[0,1].set_title('Histogram of PL shape')
axes[0,1].set_ylim(0,130)
axes[0,1].legend()

# area measurements good crystal labels using classifier trained on size + shape
hist1_output = axes[0,2].hist(area1_um2_size_shape, bins = 50, label='Area training data', alpha=0.7, color = '#79bafa', edgecolor='brown', range = (0,25))
axes[0,2].hist(area2_um2_size_shape, bins = 50, label='Area testing data', alpha=0.7,  bottom = hist1_output[0], color = '#ff9a00', edgecolor='black', range = (0,25))
axes[0,2].axvline(median_area_good_crystals_size_shape, color = '#ae0001', label = 'Median')
axes[0,2].set_xlabel('area [$\mu m^2$]')
axes[0,2].set_ylabel('count')
axes[0,2].set_title('Histogram of PL size + shape')
axes[0,2].set_ylim(0,130)
axes[0,2].legend()


# cumulative distribution function
axes[1,0].hist(area_all_um2, density=True, histtype='step', cumulative=True, label='AL', color = 'magenta')
axes[1,0].hist(area_um2, density=True, histtype='step', cumulative=True, label='PL shape', color = 'green')
axes[1,0].hist(area_um2_size_shape, density=True, histtype='step', cumulative=True, label='PL size + shape', color = 'black')
axes[1,0].grid(True)
axes[1,0].legend(loc='right')
axes[1,0].set_title('CDF of AL and PL')
axes[1,0].set_xlabel('area [$\mu m^2$]')
axes[1,0].set_ylabel('Proportion of occurence')
axes[1,0].set_xlim(0,25)

#boxplot
AL = pd.DataFrame(area_all_um2)
AL['labels'] = ['AL']*len(area_all_um2)
PL_shape = pd.DataFrame(area_um2)
PL_shape['labels'] = ['PL shape']*len(area_um2)
PL_size_shape = pd.DataFrame(area_um2_size_shape)
PL_size_shape['labels'] = ['PL size + shape']*len(area_um2_size_shape)
boxplot_table = pd.concat([AL, PL_shape,PL_size_shape], ignore_index = True)
boxplot_all = sns.boxplot(x = 'labels', y = 'area [in um^2]',  data = boxplot_table, hue = 'labels', color = '#79bafa', ax = axes[1,1])
axes[1,1].set_ylim([0,25])
axes[1,1].set_ylabel('area [$\mu m^2$]')
axes[1,1].set_title('Boxplot of AL and PL')

plt.tight_layout()
```
