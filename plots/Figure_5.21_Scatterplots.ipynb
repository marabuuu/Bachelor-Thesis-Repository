# Figure_5.21: Scatterplots

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
```

used links:
    
https://seaborn.pydata.org/tutorial/function_overview.html 

used classifier: size-shape-based classifier

```python
#all labels (after size exclusion)
table_all_both = pd.read_csv('all_measurements_both_normalized_image_um2.csv')

#only good labels
good_crystals_both = pd.read_csv('good_crystals_both_size_shape_um2.csv')
```

```python
#parameters
#all labels (after size exclusion)
area_all = table_all_both['area [in um^2]']
area = good_crystals_both['area [in um^2]']

#aspect ratio
aspect_ratio_all = table_all_both['aspect_ratio']
aspect_ratio = good_crystals_both['aspect_ratio']

#solidity
solidity_all = table_all_both['solidity']
solidity = good_crystals_both['solidity']

#all labels (after size exclusion)
circularity_all = table_all_both['circularity']
circularity = good_crystals_both['circularity']
```

```python
font = {'size' : 22}
matplotlib.rc('font',**font)
```

```python
f, axs = plt.subplots(2,2, figsize=(20, 20))

#good crystal labels area vs solidity
axs[0,0].scatter(area_all,solidity_all, label = 'DL', color = '#ffbf00')
axs[0,0].scatter(area, solidity, label = 'PL', color = '#be29ec')
axs[0,0].set_title('Area and solidity')
axs[0,0].set_xlim(-1,25)
axs[0,0].set_ylim(0.5,1)
axs[0,0].set_xlabel("area [$\mu m^2$]")
axs[0,0].set_ylabel("solidity")
axs[0,0].legend()
axs[0,0].axhline(y=0.984, color='#DED8D8', linestyle='dashed') # max solidity
axs[0,0].axvline(x=0.219, color='#DED8D8', linestyle='dashed') #min area

#good crystal labels aspect ratio vs solidity
axs[0,1].scatter(aspect_ratio_all,solidity_all, label = 'DL', color = '#ffbf00')
axs[0,1].scatter(aspect_ratio, solidity, label = 'PL', color = '#be29ec')
axs[0,1].set_title('Aspect ratio and solidity')
axs[0,1].set_xlim(0,8)
axs[0,1].set_ylim(0.5,1)
axs[0,1].set_xlabel("aspect_ratio")
axs[0,1].set_ylabel("solidity")
axs[0,1].legend()
axs[0,1].axvline(x=1.027, color='#DED8D8', linestyle='dashed') # minimum value of aspect ratio
axs[0,1].axhline(y=0.984, color='#DED8D8', linestyle='dashed') # max solidity

#good crystal labels area vs. aspect ratio
axs[1,0].scatter(area_all,circularity_all, label = 'DL', color = '#ffbf00')
axs[1,0].scatter(area, circularity, label = 'PL', color = '#be29ec')
axs[1,0].set_title('Area and circularity')
axs[1,0].set_xlim(-1,25)
axs[1,0].set_xlabel('area [$\mu m^2$]')
axs[1,0].set_ylabel('circularity')
axs[1,0].set_ylim(0,1.2)
axs[1,0].legend()
axs[1,0].axvline(x=0.219, color='#DED8D8', linestyle='dashed') #min area

#good crystal labels circularity vs. solidity
axs[1,1].scatter(circularity_all,solidity_all, label = 'DL', color = '#ffbf00')
axs[1,1].scatter(circularity, solidity, label = 'PL', color = '#be29ec')
axs[1,1].set_title('Circularity and solidity')
axs[1,1].set_xlim(0,1.2)
axs[1,1].set_ylim(0.5,1)
axs[1,1].legend()

plt.tight_layout()
```
