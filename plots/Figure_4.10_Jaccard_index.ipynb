```python
from the_segmentation_game.metrics import jaccard_index_sparse
import os
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
import pyclesperanto_prototype as cle
import numpy as np
```

```python
segmentation_folder = '../crystal data/3. segmentation selection for jaccard index/'
manual_folder = '../crystal data/3. manual segmentation/'
```

```python
file_list = sorted(os.listdir(segmentation_folder))
file_list
```

```python
jaccard_list = []
for i, filename in enumerate(file_list):
    fig, axs = plt.subplots(1, 2, figsize=(25,25))
    
    #show original
    print(filename)
    manual = imread(manual_folder + filename)
    cle.imshow(manual, labels = True, plot=axs[0])
    
     #show result of object segmenter
    segmentation = imread(segmentation_folder + filename)
    cle.imshow(segmentation, labels = True, plot=axs[1])
    
    #compute jaccard index
    compute_jaccard = jaccard_index_sparse(manual, segmentation)
    jaccard_list.append(compute_jaccard)
    
    #show all images
    plt.show()
```

```python
#mean of list of jaccard indices
np.mean(jaccard_list)
```
