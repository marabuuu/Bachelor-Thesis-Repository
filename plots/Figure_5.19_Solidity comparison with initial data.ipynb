# Figure_5.19: Solidity comparison with initial data

```python
import os
import matplotlib.pyplot as plt
from skimage.io import imread
import pyclesperanto_prototype as cle
import matplotlib

cle.select_device("Intel(R) Iris(TM) Plus Graphics 640")
```

```python
font = {'size' : 22}
matplotlib.rc('font',**font)
```

```python
segmentation = cle.push(imread('/Users/mara/Desktop/Bachelorarbeit/Robert - PoL/Solidity/size + shape/2. segmentation with solidity/img09_testing.tif'))
segmentation_crop = segmentation[200:600, 600:1000]
solidity = cle.push(imread('/Users/mara/Desktop/Bachelorarbeit/Robert - PoL/Solidity/size + shape/4. solidity/img09_testing.tif'))
solidity_crop = solidity[200:600, 600:1000]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20,15))
cle.imshow(segmentation_crop, plot=axs[0], labels = False, colormap = 'plasma', min_display_intensity = 0.5)
axs[0].set_title('Solidity of AL')
cle.imshow(solidity_crop, plot=axs[1], labels = False, colormap = 'plasma', min_display_intensity = 0.5)
axs[1].set_title('Solidity of PL')
```
