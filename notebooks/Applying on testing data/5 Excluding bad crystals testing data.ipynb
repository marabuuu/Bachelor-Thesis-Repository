# Excluding bad crystals testing data 

```python
import os
from skimage.io import imread, imsave
import numpy as np
from pyclesperanto_prototype import imshow
```

```python
label_folder = "/Users/mara/Desktop/Bachelorarbeit/Robert - PoL/20220523_training data/no training data/3. in range result of object segmenter/"
classifier_result_folder = "/Users/mara/Desktop/Bachelorarbeit/Robert - PoL/20220523_training data/no training data/4. result of object classifier training data/"
```

```python
file_list = sorted(os.listdir(label_folder))
file_list
```

```python
label_list = []
classification_result_list = []
```

```python
for i, filename in enumerate(file_list):
    label = imread(label_folder + filename)
    classification = imread(classifier_result_folder + filename)
    label_list.append(label)
    classification_result_list.append(classification)
```

```python
good_crystal_labels = "/Users/mara/Desktop/Bachelorarbeit/Robert - PoL/20220523_training data/no training data/5. good crystal labels testing data/"
```

```python
os.makedirs(good_crystal_labels, exist_ok=True)
```

```python
for class_image,label,filename in zip(classification_result_list,label_list,file_list):
    class_image = np.asarray(class_image)
    class_image[class_image == 1]=0
    class_image_mask = class_image.astype(bool)
    label_image_filtered=np.copy(label)
    label_image_filtered[class_image_mask==False]=0
    print(filename)
    imshow(label_image_filtered, colorbar= True,colormap='jet',min_display_intensity=0)
    imsave(good_crystal_labels+filename,label_image_filtered)
```
